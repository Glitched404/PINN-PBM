```python
### Pre Final Run for case 1 and 2 (also checking for case 3 and 4 ) ###
## Enhanced PINN with Adaptive Physics Loss Scaling ##
import os
import time
import random
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tqdm import tqdm
from tensorflow.keras.regularizers import l2

# ================================================================
# TensorFlow Probability for L-BFGS
# ================================================================
try:
    import tensorflow_probability as tfp
    tfp_opt = tfp.optimizer
except Exception:
    tfp = None
    tfp_opt = None

# ================================================================
# GPU setup
# ================================================================
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    for gpu in gpus:
        tf.config.experimental.set_memory_growth(gpu, True)
    tf.config.optimizer.set_jit(False)

# ================================================================
# Reproducibility
# ================================================================
seed = 42
os.environ['PYTHONHASHSEED'] = str(seed)
random.seed(seed)
np.random.seed(seed)
tf.random.set_seed(seed)

# ================================================================
# Helper Functions
# ================================================================
@tf.function
def trapz_tf(y, x):
    n = tf.size(x)
    def no_integration():
        return tf.constant(0.0, dtype=tf.float32)
    def do_integration():
        return tf.reduce_sum(0.5 * (y[:-1] + y[1:]) * (x[1:] - x[:-1]))
    return tf.cond(n < 2, no_integration, do_integration)

@tf.function
def huber_loss(residuals, delta=0.1):
    """Huber loss: quadratic for small residuals, linear for large ones"""
    abs_res = tf.abs(residuals)
    quadratic = tf.minimum(abs_res, delta)
    linear = abs_res - quadratic
    return 0.5 * quadratic**2 + delta * linear

@tf.function
def percentile_clip_loss(residuals, weights, percentile=95.0):
    """Ignore worst residuals by clipping to percentile"""
    weighted_sq = weights * tf.square(residuals)
    threshold = tfp.stats.percentile(weighted_sq, percentile)
    clipped = tf.minimum(weighted_sq, threshold)
    return tf.reduce_mean(clipped)

def delta_peak(x_grid, center, area, width_fraction=0.01):
    x_grid_np = np.array(x_grid)
    width = center * width_fraction
    if width < 1e-9:
        positive_x = x_grid_np[x_grid_np > 0]
        width = np.min(positive_x) * width_fraction if len(positive_x) > 0 else 1e-9
    height = area / (width * np.sqrt(2 * np.pi) + 1e-30)
    return height * np.exp(-0.5 * ((x_grid_np - center) / width) ** 2)

def analytic_f(v, t, case_type):
    if case_type == 'case1':
        return np.exp(-v * (1.0 + t)) * (1.0 + t) ** 2 + 1e-30
    elif case_type == 'case2':
        return np.exp(-t * v ** 2 - v) * (1.0 + 2.0 * t * (1.0 + v)) + 1e-30
    elif case_type == 'case3':
        Rx = 1.0
        heaviside = (v < Rx).astype(float)
        area_delta = np.exp(-t * Rx)
        delta_part = delta_peak(v, Rx, area_delta)
        continuous_part = np.exp(-t * v) * (2.0 * t + t ** 2 * (Rx - v)) * heaviside
        return delta_part + continuous_part + 1e-30
    elif case_type == 'case4':
        Rx = 1.0
        heaviside = (v < Rx).astype(float)
        area_delta = np.exp(-t * Rx ** 2)
        delta_part = delta_peak(v, Rx, area_delta)
        continuous_part = np.exp(-t * v ** 2) * (2.0 * t * Rx) * heaviside
        return delta_part + continuous_part + 1e-30
    else:
        raise ValueError(f"Unsupported case type: {case_type}")

def get_dynamic_loss_weight(epoch, total_epochs, final_weight, burn_in_fraction=0.3, ramp_up_fraction=0.6):
    burn_in_epochs = int(total_epochs * burn_in_fraction)
    ramp_up_epochs = int(total_epochs * ramp_up_fraction)
    if epoch < burn_in_epochs:
        return 0.0
    elif epoch < burn_in_epochs + ramp_up_epochs:
        progress = (epoch - burn_in_epochs) / ramp_up_epochs
        return final_weight * np.sqrt(progress)
    else:
        return final_weight

def get_custom_dynamic_loss_weight(epoch, total_epochs):
    stage1_end_epoch = 1000
    stage2_end_epoch = 3000
    if epoch < stage1_end_epoch:
        return 0.0
    elif epoch < stage2_end_epoch:
        start_weight, end_weight = 0.0, 1.0
        progress = (epoch - stage1_end_epoch) / (stage2_end_epoch - stage1_end_epoch)
        return start_weight + progress * (end_weight - start_weight)
    else:
        return 1.0

@tf.function
def _tail_trapz(g, v_grid):
    dv = v_grid[1:] - v_grid[:-1]
    g_avg = 0.5 * (g[..., :-1] + g[..., 1:])
    seg = g_avg * dv
    rev = tf.reverse(seg, axis=[-1])
    rev_cumsum = tf.cumsum(rev, axis=-1)
    rev_cum = tf.reverse(rev_cumsum, axis=[-1])
    TI = tf.concat([rev_cum, tf.zeros_like(rev_cum[..., :1])], axis=-1)
    return TI

# ================================================================
# BreakagePINN Class with Enhanced Loss Scaling
# ================================================================
class BreakagePINN:
    def __init__(self, v_min, v_max, t_min, t_max, n_v=500, n_t=100,
                 n_hidden_layers=4, n_neurons=40, case_type='case1',
                 initial_learning_rate=2e-4, epochs=5000,
                 loss_scaling='adaptive_huber'):
        """
        loss_scaling options:
        - 'original': Your original weighting
        - 'huber': Huber loss with adaptive weighting
        - 'adaptive_epsilon': Adaptive epsilon based on batch statistics
        - 'adaptive_huber': Combines adaptive epsilon + Huber (RECOMMENDED)
        - 'percentile': Percentile-based outlier rejection
        """
        self.v_min, self.v_max = v_min, v_max
        self.t_min, self.t_max = t_min, t_max
        self.case_type = case_type
        self.loss_scaling = loss_scaling

        self.log_vmin = tf.constant(np.log(v_min), dtype=tf.float32)
        self.log_vmax = tf.constant(np.log(v_max), dtype=tf.float32)
        self.t_min_tensor = tf.constant(t_min, dtype=tf.float32)
        self.log_t_min_tensor = tf.constant(np.log(t_min + 1e-8), dtype=tf.float32)
        self.log_t_max_tensor = tf.constant(np.log(t_max), dtype=tf.float32)

        ratio = 2.0 ** (1.0 / 3.0)
        self.v_grid_tf = tf.constant((v_min * ratio ** np.arange(n_v)).astype(np.float32))
        self.t_grid = np.linspace(t_min, t_max, n_t, dtype=np.float32)

        self.model = self.build_model(n_hidden_layers, n_neurons)
        self.amt = None
        if self.case_type in ['case3', 'case4']:
            self.build_amt_head(n_hidden=2, width=32)

        # Use warmup + cosine decay for better convergence
        warmup_steps = 500
        total_steps = epochs
        self.lr_schedule = tf.keras.optimizers.schedules.CosineDecay(
            initial_learning_rate=initial_learning_rate,
            decay_steps=total_steps - warmup_steps,
            alpha=0.01  # Min LR = 1% of initial
        )

        # Wrap with warmup
        class WarmupSchedule(tf.keras.optimizers.schedules.LearningRateSchedule):
            def __init__(self, base_schedule, warmup_steps, initial_lr):
                self.base_schedule = base_schedule
                self.warmup_steps = tf.cast(warmup_steps, tf.float32)
                self.initial_lr = initial_lr

            def __call__(self, step):
                step = tf.cast(step, tf.float32)
                warmup_lr = self.initial_lr * (step / self.warmup_steps)
                return tf.cond(
                    step < self.warmup_steps,
                    lambda: warmup_lr,
                    lambda: self.base_schedule(step - self.warmup_steps)
                )

        self.lr_schedule = WarmupSchedule(self.lr_schedule, warmup_steps, initial_learning_rate)
        self.optimizer = tf.keras.optimizers.Adam(learning_rate=self.lr_schedule, clipnorm=1.0)

        self.train_loss_history, self.data_loss_history, self.physics_loss_history = [], [], []

        # Track statistics for adaptive methods
        self.f_magnitude_history = []

    def build_model(self, n_hidden_layers, n_neurons):
        inputs = tf.keras.Input(shape=(2,), dtype=tf.float32)
        x = tf.keras.layers.Dense(n_neurons, activation='tanh',
                                  kernel_initializer='glorot_normal',
                                  kernel_regularizer=l2(1e-6))(inputs)
        for i in range(n_hidden_layers - 1):
            if (i % 2) == 0:
                residual = x
            x_inner = tf.keras.layers.Dense(n_neurons, activation='tanh',
                                            kernel_initializer='glorot_normal',
                                            kernel_regularizer=l2(1e-6))(x)
            if (i % 2) != 0:
                x = tf.keras.layers.Add()([x_inner, residual])
            else:
                x = x_inner
        outputs = tf.keras.layers.Dense(1, activation=None,
                                        kernel_regularizer=l2(1e-6))(x)
        return tf.keras.Model(inputs, outputs)

    def build_amt_head(self, n_hidden=2, width=32):
        t_in = tf.keras.Input(shape=(1,), dtype=tf.float32)
        x = t_in
        for _ in range(n_hidden):
            x = tf.keras.layers.Dense(width, activation='tanh',
                                      kernel_initializer='glorot_normal',
                                      kernel_regularizer=l2(1e-6))(x)
        out = tf.keras.layers.Dense(1, activation='softplus',
                                    kernel_regularizer=l2(1e-6))(x)
        self.amt = tf.keras.Model(t_in, out)

    def net_logf(self, v, t):
        log_v = tf.math.log(v)
        log_t = tf.math.log(t + 1e-8)
        xi = 2.0 * (log_v - self.log_vmin) / (self.log_vmax - self.log_vmin) - 1.0
        tau_scaled = 2.0 * (log_t - self.log_t_min_tensor) / (self.log_t_max_tensor - self.log_t_min_tensor) - 1.0
        vt = tf.stack([xi, tau_scaled], axis=1)
        return tf.squeeze(self.model(vt), axis=1)

    def net_f(self, v, t):
        logf = self.net_logf(v, t)
        logf = tf.clip_by_value(logf, -30.0, 30.0)
        f_cont = tf.exp(logf)
        if self.case_type in ['case3', 'case4']:
            Rx = tf.constant(1.0, tf.float32)
            f_cont = tf.where(v < Rx, f_cont, tf.zeros_like(f_cont))
            sigma = 0.01 * Rx
            A = tf.squeeze(self.amt(tf.expand_dims(t, 1)), axis=1) if self.amt is not None else tf.zeros_like(t)
            spike = A * tf.exp(-0.5 * tf.square((v - Rx) / sigma)) / (sigma * tf.sqrt(2.0 * np.pi))
            return f_cont + spike
        return f_cont

    def selection_function(self, v):
        if self.case_type in ['case1', 'case3']:
            return v
        elif self.case_type in ['case2', 'case4']:
            return tf.square(v)
        else:
            raise ValueError(f"Unsupported case type: {self.case_type}")

    def breakage_distribution(self, v, vp):
        return 2.0 / vp

    @tf.function
    def compute_pointwise_residuals_vectorized(self, v_c, t_c):
        v_grid = self.v_grid_tf
        B = tf.shape(v_c)[0]
        N = tf.shape(v_grid)[0]

        with tf.GradientTape() as tape:
            tape.watch(t_c)
            f_c = self.net_f(v_c, t_c)
        df_dt = tape.gradient(f_c, t_c)
        df_dt = tf.where(tf.math.is_finite(df_dt), df_dt, tf.zeros_like(df_dt))

        death = - self.selection_function(v_c) * f_c

        t_grid_batch = tf.repeat(tf.expand_dims(t_c, 1), N, axis=1)
        v_grid_batch = tf.repeat(tf.expand_dims(v_grid, 0), B, axis=0)
        f_grid = self.net_f(tf.reshape(v_grid_batch, [-1]),
                            tf.reshape(t_grid_batch, [-1]))
        f_grid = tf.reshape(f_grid, [B, N])

        s_v = self.selection_function(v_grid)
        g_grid = f_grid * (2.0 / v_grid) * s_v

        TI = _tail_trapz(g_grid, v_grid)

        idx = tf.searchsorted(v_grid, v_c, side='right')
        idx = tf.minimum(idx, N - 1)
        batch_ids = tf.range(B, dtype=idx.dtype)
        birth = tf.gather_nd(TI, tf.stack([batch_ids, idx], axis=1))

        return df_dt - (birth + death), f_c

    @tf.function
    def compute_physics_loss(self, v_colloc, t_colloc):
        raw_residuals, f_pred = self.compute_pointwise_residuals_vectorized(v_colloc, t_colloc)

        # Clip residuals to prevent explosion
        raw_residuals = tf.clip_by_value(raw_residuals, -1e3, 1e3)

        if self.loss_scaling == 'original':
            # Original approach
            weights = 1.0 / (tf.square(f_pred) + 1e-6)
            weights = tf.minimum(weights, 1e2)
            residual_loss = tf.reduce_mean(weights * tf.square(raw_residuals))

        elif self.loss_scaling == 'huber':
            # Huber loss with original weighting
            weights = 1.0 / (tf.square(f_pred) + 1e-6)
            weights = tf.minimum(weights, 1e2)
            residual_loss = tf.reduce_mean(weights * huber_loss(raw_residuals, delta=0.1))

        elif self.loss_scaling == 'adaptive_epsilon':
            # Adaptive epsilon based on batch statistics
            f_sq = tf.square(f_pred)
            median_f_sq = tfp.stats.percentile(f_sq, 50.0)
            adaptive_eps = 0.01 * median_f_sq + 1e-10
            weights = 1.0 / (f_sq + adaptive_eps)
            weights = tf.minimum(weights, 1e2)
            residual_loss = tf.reduce_mean(weights * tf.square(raw_residuals))

        elif self.loss_scaling == 'adaptive_huber':
            # RECOMMENDED: Combines adaptive epsilon + Huber loss
            f_sq = tf.square(f_pred)
            median_f_sq = tfp.stats.percentile(f_sq, 50.0)
            adaptive_eps = 0.01 * median_f_sq + 1e-10
            weights = 1.0 / (f_sq + adaptive_eps)
            weights = tf.minimum(weights, 50.0)  # Less aggressive clipping
            residual_loss = tf.reduce_mean(weights * huber_loss(raw_residuals, delta=0.1))

        elif self.loss_scaling == 'percentile':
            # Percentile-based outlier rejection
            weights = 1.0 / (tf.square(f_pred) + 1e-6)
            weights = tf.minimum(weights, 1e2)
            residual_loss = percentile_clip_loss(raw_residuals, weights, percentile=95.0)

        else:
            raise ValueError(f"Unknown loss_scaling: {self.loss_scaling}")

        residual_loss = tf.clip_by_value(residual_loss, 0.0, 1e6)
        return residual_loss

    @tf.function
    def compute_data_loss(self, v_obs, t_obs, f_obs):
        log_f_obs = tf.math.log(tf.squeeze(f_obs, axis=1) + 1e-12)
        log_f_pred = self.net_logf(v_obs, t_obs)

        log_f_pred = tf.clip_by_value(log_f_pred, -20.0, 20.0)
        log_f_obs = tf.clip_by_value(log_f_obs, -20.0, 20.0)

        # Adaptive weighting for data loss too
        log_f_sq = tf.square(log_f_obs)
        median_log_f_sq = tfp.stats.percentile(log_f_sq, 50.0)
        adaptive_eps = 0.01 * median_log_f_sq + 1e-8
        weights = 1.0 / (log_f_sq + adaptive_eps)
        weights = tf.minimum(weights, 50.0)

        return tf.reduce_mean(weights * tf.square(log_f_pred - log_f_obs))

    @tf.function
    def train_step(self, v_d, t_d, f_d, v_p, t_p, w_data, w_phys):
        train_vars = self.model.trainable_variables
        if self.amt is not None:
            train_vars = train_vars + self.amt.trainable_variables
        with tf.GradientTape() as tape:
            data_l = self.compute_data_loss(v_d, t_d, f_d)
            phys_l = self.compute_physics_loss(v_p, t_p)
            total_l = w_data * data_l + w_phys * phys_l
        grads = tape.gradient(total_l, train_vars)
        grads_vars = tuple((g, v) for g, v in zip(grads, train_vars) if g is not None)
        self.optimizer.apply_gradients(grads_vars)
        return total_l, data_l, phys_l

    def train(self, v_data, t_data, f_data, v_cand, t_cand,
              epochs=2000, data_batch_size=512, phys_batch_size=512,
              phys_loss_weight=0.01, rar_refinement_interval=2000,
              loss_schedule='fixed'):
        num_data = v_data.shape[0]
        v_cand = tf.convert_to_tensor(v_cand, tf.float32)
        t_cand = tf.convert_to_tensor(t_cand, tf.float32)

        v_p_refined = tf.constant([], dtype=tf.float32)
        t_p_refined = tf.constant([], dtype=tf.float32)

        t_min_eff_global = max(self.t_min + 1e-6, 1e-3 * self.t_max)

        for epoch in tqdm(range(epochs), desc=f"ADAM Training ({self.loss_scaling})"):
            if loss_schedule == 'dynamic':
                current_phys_weight = get_dynamic_loss_weight(epoch, epochs, phys_loss_weight)
            elif loss_schedule == 'custom':
                current_phys_weight = get_custom_dynamic_loss_weight(epoch, epochs)
            else:
                current_phys_weight = phys_loss_weight

            if epoch > 0 and epoch % rar_refinement_interval == 0:
                all_scores = []
                rar_batch_size = 8192
                num_cand_pts = v_cand.shape[0]
                for i in range(0, num_cand_pts, rar_batch_size):
                    v_cand_batch = v_cand[i:i+rar_batch_size]
                    t_cand_batch = t_cand[i:i+rar_batch_size]
                    res_cand, f_cand = self.compute_pointwise_residuals_vectorized(v_cand_batch, t_cand_batch)
                    rel_score = tf.square(res_cand) / (tf.square(f_cand) + 1e-10)
                    rel_score = tf.minimum(rel_score, 1e6)
                    all_scores.append(rel_score)
                full_rel_score = tf.concat(all_scores, axis=0)
                k_refine = max(phys_batch_size // 2, 64)
                _, top_idx = tf.math.top_k(full_rel_score, k=k_refine)
                v_p_refined = tf.gather(v_cand, top_idx)
                t_p_refined = tf.gather(t_cand, top_idx)

            idx = np.random.choice(num_data, data_batch_size, replace=True)
            v_d, t_d, f_d = v_data[idx], t_data[idx], f_data[idx]
            ic_every = 1
            if (epoch % ic_every) == 0:
                ic_bs = max(16, data_batch_size // 2)
                t_ic = np.full((ic_bs,), self.t_min, np.float32)
                v_ic = np.exp(np.random.uniform(np.log(self.v_min), np.log(self.v_max), ic_bs)).astype(np.float32)
                f_ic = analytic_f(v_ic, self.t_min, self.case_type).astype(np.float32).reshape(-1, 1)
                v_d = np.concatenate([v_d, v_ic], axis=0)
                t_d = np.concatenate([t_d, t_ic], axis=0)
                f_d = np.concatenate([f_d, f_ic], axis=0)

            n_refined = int(v_p_refined.shape[0])
            n_tail = phys_batch_size // 8
            n_random = max(phys_batch_size - n_refined - n_tail, 0)
            v_tail_thresh = self.v_max * 0.2
            v_p_tail = np.exp(np.random.uniform(np.log(v_tail_thresh), np.log(self.v_max), n_tail)).astype(np.float32)
            t_p_tail = np.random.uniform(t_min_eff_global, self.t_max, n_tail).astype(np.float32)
            v_p_rand = np.exp(np.random.uniform(np.log(self.v_min), np.log(self.v_max), n_random)).astype(np.float32)
            t_p_rand = np.random.uniform(t_min_eff_global, self.t_max, n_random).astype(np.float32)
            v_p = tf.concat([v_p_rand, v_p_tail, v_p_refined], axis=0)
            t_p = tf.concat([t_p_rand, t_p_tail, t_p_refined], axis=0)

            L, Ld, Lp = self.train_step(v_d, t_d, f_d, v_p, t_p, 1.0,
                                        tf.constant(current_phys_weight, dtype=tf.float32))

            self.train_loss_history.append(float(L.numpy()))
            self.data_loss_history.append(float(Ld.numpy()))
            self.physics_loss_history.append(float(Lp.numpy()))

            if epoch % 500 == 0:
                lr = float(self.lr_schedule(self.optimizer.iterations).numpy())
                print(f"\nEpoch {epoch}: total={L:.3e}, data={Ld:.3e}, phys={Lp:.3e}, w_phys={current_phys_weight:.4f}, lr={lr:.3e}")

    def predict(self, v_pts, t_pts):
        V, T = np.meshgrid(v_pts, t_pts)
        f_pred = self.net_f(tf.constant(V.flatten(), tf.float32),
                            tf.constant(T.flatten(), tf.float32))
        return f_pred.numpy().reshape(V.shape)

# ================================================================
# L-BFGS Optimizers
# ================================================================
from scipy.optimize import minimize

def lbfgs_optimizer_scipy(pinn, v_train_tf, t_train_tf, f_train_tf, v_colloc_tf, t_colloc_tf, max_iter=1000):
    train_vars = pinn.model.trainable_variables
    if pinn.amt is not None:
        train_vars = train_vars + pinn.amt.trainable_variables

    shapes = [v.shape.as_list() for v in train_vars]
    w0 = np.concatenate([v.numpy().flatten() for v in train_vars]).astype(np.float64)

    def set_weights(w_flat):
        idx = 0
        for var, shape in zip(train_vars, shapes):
            size = int(np.prod(shape))
            var.assign(tf.reshape(tf.convert_to_tensor(w_flat[idx:idx+size], dtype=tf.float32), shape))
            idx += size

    def loss_and_grad(w_flat):
        set_weights(w_flat)
        with tf.GradientTape() as tape:
            data_l = pinn.compute_data_loss(v_train_tf, t_train_tf, f_train_tf)
            phys_l = pinn.compute_physics_loss(v_colloc_tf, t_colloc_tf)
            total_loss = data_l + phys_l
        grads = tape.gradient(total_loss, train_vars)
        safe_grads = [(g if g is not None else tf.zeros_like(v)) for g, v in zip(grads, train_vars)]
        grad_flat = np.concatenate([g.numpy().flatten() for g in safe_grads]).astype(np.float64)
        return float(total_loss.numpy().astype(np.float64)), grad_flat

    pbar = tqdm(total=max_iter, desc="L-BFGS fine-tuning")

    def lbfgs_cb(xk):
        data_l = pinn.compute_data_loss(v_train_tf, t_train_tf, f_train_tf)
        phys_l = pinn.compute_physics_loss(v_colloc_tf, t_colloc_tf)
        total_l = data_l + phys_l
        pinn.train_loss_history.append(float(total_l.numpy()))
        pinn.data_loss_history.append(float(data_l.numpy()))
        pinn.physics_loss_history.append(float(phys_l.numpy()))
        pbar.update(1)
        pbar.set_postfix(loss=f"{float(total_l.numpy()):.3e}")

    res = minimize(
        fun=loss_and_grad, x0=w0, jac=True, method='L-BFGS-B',
        callback=lbfgs_cb,
        options={'maxiter': max_iter, 'disp': False, 'gtol': 0.0, 'ftol': 0.0, 'maxfun': int(1e9)}
    )
    pbar.close()
    set_weights(res.x)
    return res

def lbfgs_optimizer_tfp(pinn, v_train_tf, t_train_tf, f_train_tf, v_colloc_tf, t_colloc_tf, max_iter=1500):
    if tfp_opt is None:
        raise RuntimeError("TensorFlow Probability not available.")

    # Collect all trainable variables from the PINN model
    var_list = list(pinn.model.trainable_variables)
    if pinn.amt is not None:
        var_list += list(pinn.amt.trainable_variables)

    # Compute shapes and sizes for flattening/unflattening variables
    shapes = [v.shape for v in var_list]
    sizes = [int(np.prod(s)) for s in shapes]
    idxs = np.cumsum([0] + sizes)

    # Helper function to flatten variable list into 1D vector
    def pack(vars_): return tf.concat([tf.reshape(v, [-1]) for v in vars_], axis=0)

    # Helper function to unflatten 1D vector back into variable shapes
    def unpack(w): return [tf.reshape(w[idxs[i]:idxs[i+1]], shapes[i]) for i in range(len(var_list))]

    # REMOVED: Progress bar initialization (tqdm)
    # REMOVED: iteration_count tracking list

    @tf.function
    def value_and_gradients(w):
        # Unpack flattened parameters and assign to model variables
        ws = unpack(w)
        for var, new_val in zip(var_list, ws):
            var.assign(tf.cast(new_val, var.dtype))

        # Compute losses with gradient tape
        with tf.GradientTape(persistent=True) as tape:
            # Compute data loss and clamp non-finite values
            data_l = pinn.compute_data_loss(v_train_tf, t_train_tf, f_train_tf)
            data_l = tf.where(tf.math.is_finite(data_l), data_l, tf.constant(1e10, dtype=tf.float32))

            # Compute physics loss and clamp non-finite values
            phys_l = pinn.compute_physics_loss(v_colloc_tf, t_colloc_tf)
            phys_l = tf.where(tf.math.is_finite(phys_l), phys_l, tf.constant(1e10, dtype=tf.float32))

            # Add L2 weight decay regularization
            weight_decay = 1e-6 * tf.add_n([tf.reduce_sum(tf.square(v)) for v in var_list])

            # Combine all loss components
            total_loss = data_l + phys_l + weight_decay

        # Compute gradients for all trainable variables
        grads = tape.gradient(total_loss, var_list)
        del tape

        # Ensure all gradients are safe (replace None or non-finite with zeros)
        safe_grads = []
        for g, v in zip(grads, var_list):
            if g is None:
                g = tf.zeros_like(v)
            else:
                g = tf.where(tf.math.is_finite(g), g, tf.zeros_like(g))
            safe_grads.append(g)

        # Flatten gradient list into 1D vector
        grad_vector = pack(safe_grads)
        grad_vector = tf.where(tf.math.is_finite(grad_vector), grad_vector, tf.zeros_like(grad_vector))

        # Return loss and gradient with consistent dtype
        return tf.cast(total_loss, w.dtype), tf.cast(grad_vector, w.dtype)

    # REMOVED: value_and_gradients_with_progress wrapper function
    # REMOVED: Progress bar updates
    # REMOVED: Loss history appending via numpy conversion

    # Flatten initial parameters into 1D vector
    w0 = pack(var_list)

    print(f"\nStarting TFP L-BFGS with {len(w0)} parameters...")

    # Run L-BFGS optimization (now directly using value_and_gradients)
    results = tfp_opt.lbfgs_minimize(
        value_and_gradients_function=value_and_gradients,  # Direct function without progress wrapper
        initial_position=w0,
        max_iterations=max_iter,
        tolerance=1e-7,
        f_relative_tolerance=1e-9,
        x_tolerance=1e-12,
        initial_inverse_hessian_estimate=None,
        max_line_search_iterations=30,
        parallel_iterations=4
    )

    # REMOVED: pbar.close()

    # Unpack final optimized parameters and assign back to model variables
    final_vars = unpack(results.position)
    for var, new_val in zip(var_list, final_vars):
        var.assign(tf.cast(new_val, var.dtype))

    print(f"L-BFGS Complete: Converged={bool(results.converged.numpy())}, Iterations={int(results.num_iterations.numpy())}")

    return results


# ================================================================
# Main Runner
# ================================================================
def run_test_case(case_type='case1', optimizer='tfp', loss_scaling='adaptive_huber'):
    """
    Run test case with specified loss scaling strategy.

    loss_scaling options:
    - 'original': Original approach (baseline)
    - 'huber': Huber loss with original weighting
    - 'adaptive_epsilon': Adaptive epsilon based on batch statistics
    - 'adaptive_huber': Adaptive epsilon + Huber (RECOMMENDED)
    - 'percentile': Percentile-based outlier rejection
    """
    start_time = time.time()

    print(f"\n{'='*60}")
    print(f"Running Test Case: {case_type.upper()}")
    print(f"Optimizer: {optimizer.upper()}")
    print(f"Loss Scaling: {loss_scaling.upper()}")
    print(f"{'='*60}\n")

    params = {
        'case1': {'v_min': 1e-3, 'v_max': 1e1, 't_max': 10.0, 't_slices': [0.0, 2.0, 5.0, 10.0],
                  'epochs': 3000, 'lbfgs_iter': 2500, 'phys_loss_weight': 0.01, 'lr': 5e-4,
                  'ylim': (1e-1, 150), 'loss_schedule': 'dynamic'},
        'case2': {'v_min': 1e-3, 'v_max': 1e2, 't_max': 5.0, 't_slices': [0.0, 1.0, 2.0, 5.0],
                  'epochs': 3000, 'lbfgs_iter': 3000, 'phys_loss_weight': 0.01, 'lr': 1e-4,
                  'ylim': (1e-10, 15), 'loss_schedule': 'dynamic'},
        'case3': {'v_min': 1e-4, 'v_max': 10e-1, 't_max': 1000.0, 't_slices': [0.0, 200.0, 500.0, 1000.0],
                  'epochs': 5000, 'lbfgs_iter': 3000, 'phys_loss_weight': 0.01, 'lr': 1e-4,
                  'ylim': (1e-3, 12e5), 'loss_schedule': 'dynamic'},
        'case4': {'v_min': 1e-4, 'v_max': 1.5, 't_max': 2000.0, 't_slices': [0.0, 500.0, 1000.0, 2000.0],
                  'epochs': 5000, 'lbfgs_iter': 3000, 'phys_loss_weight': 0.01, 'lr': 1e-4,
                  'ylim': (1e-2, 5e3), 'loss_schedule': 'dynamic'},
    }[case_type]

    # Training grid
    ratio = 2.0 ** (1.0 / 3.0)
    num_v_pts = int(np.ceil(np.log(params['v_max'] / params['v_min']) / np.log(ratio))) + 1
    v_grid_train = (params['v_min'] * ratio ** np.arange(num_v_pts)).astype(np.float32)

    vs, ts, fs = [], [], []
    for t_val in params['t_slices']:
        f_vals = analytic_f(v_grid_train, t_val, case_type)
        for i, v_val in enumerate(v_grid_train):
            vs.append(v_val); ts.append(t_val); fs.append(f_vals[i])

    v_train = np.array(vs, dtype=np.float32)
    t_train = np.array(ts, dtype=np.float32)
    f_train = np.array(fs, dtype=np.float32).reshape(-1, 1)

    pinn = BreakagePINN(params['v_min'], params['v_max'], 0.0, params['t_max'],
                        n_v=num_v_pts, n_t=len(params['t_slices']),
                        n_hidden_layers=8, n_neurons=128,
                        case_type=case_type, initial_learning_rate=params['lr'],
                        epochs=params['epochs'], loss_scaling=loss_scaling)

    # Candidate physics grid for RAR
    v_cand_flat = np.logspace(np.log10(params['v_min']), np.log10(params['v_max']), 1000).astype(np.float32)
    t_cand_flat = np.linspace(0.0, params['t_max'], 200).astype(np.float32)
    V_cand, T_cand = np.meshgrid(v_cand_flat, t_cand_flat)

    # ADAM training with larger batch sizes
    pinn.train(v_train, t_train, f_train, V_cand.flatten(), T_cand.flatten(),
               epochs=params['epochs'],
               data_batch_size=1024,  # Increased from 512
               phys_batch_size=2048,  # Increased from 512
               phys_loss_weight=params['phys_loss_weight'],
               rar_refinement_interval=5000,
               loss_schedule=params['loss_schedule'])

    # L-BFGS fine-tuning with more collocation points
    n_colloc_lbfgs = 2048  # Increased from 512
    v_p_lbfgs = np.exp(np.random.uniform(np.log(pinn.v_min), np.log(pinn.v_max), n_colloc_lbfgs)).astype(np.float32)
    t_p_lbfgs = np.random.uniform(pinn.t_min, pinn.t_max, n_colloc_lbfgs).astype(np.float32)

    v_train_tf = tf.convert_to_tensor(v_train, tf.float32)
    t_train_tf = tf.convert_to_tensor(t_train, tf.float32)
    f_train_tf = tf.convert_to_tensor(f_train, tf.float32)
    v_colloc_tf = tf.convert_to_tensor(v_p_lbfgs, tf.float32)
    t_colloc_tf = tf.convert_to_tensor(t_p_lbfgs, tf.float32)

    if optimizer == 'tfp':
        results = lbfgs_optimizer_tfp(pinn, v_train_tf, t_train_tf, f_train_tf, v_colloc_tf, t_colloc_tf,
                                      max_iter=params['lbfgs_iter'])
    else:
        _ = lbfgs_optimizer_scipy(pinn, v_train_tf, t_train_tf, f_train_tf, v_colloc_tf, t_colloc_tf,
                                  max_iter=params['lbfgs_iter'])

    # Final results
    v_plot = np.logspace(np.log10(params['v_min']), np.log10(params['v_max']), 500)
    t_plot = np.array(params['t_slices'], dtype=np.float32)
    f_pred = pinn.predict(v_plot, t_plot)
    f_exact = np.array([analytic_f(v_plot, t, case_type) for t in t_plot])

    # Calculate relative error
    relative_errors = []
    for i in range(len(t_plot)):
        rel_err = np.abs((f_pred[i] - f_exact[i]) / (f_exact[i] + 1e-30))
        mean_rel_err = np.mean(rel_err)
        relative_errors.append(mean_rel_err)

    end_time = time.time()
    total_time = end_time - start_time

    # Print final summary
    print(f"\n{'='*60}")
    print(f"FINAL RESULTS - {case_type.upper()} ({loss_scaling})")
    print(f"{'='*60}")
    print(f"Total training time: {total_time:.2f} seconds ({total_time/60:.2f} minutes)")
    print(f"Final data loss: {pinn.data_loss_history[-1]:.3e}")
    print(f"Final physics loss: {pinn.physics_loss_history[-1]:.3e}")
    print(f"Final total loss: {pinn.train_loss_history[-1]:.3e}")
    print(f"\nRelative errors at each time slice:")
    for i, t in enumerate(t_plot):
        print(f"  t={t:.1f}: {relative_errors[i]:.3e}")
    print(f"Mean relative error: {np.mean(relative_errors):.3e}")
    print(f"{'='*60}\n")

    # Plots
    plt.figure(figsize=(12, 9))
    markers = ['o', 's', '^', 'd', 'x', '+']
    for i, t in enumerate(t_plot):
        plt.semilogx(v_plot, f_exact[i], '-', label=f'Analytic t={t:.1f}')
        plt.semilogx(v_plot[::20], f_pred[i][::20], markers[i % len(markers)], linestyle='None',
                     markerfacecolor='none', markersize=6, markeredgewidth=1, label=f'PINN t={t:.1f}')
    plt.xlim(params['v_min'], params['v_max'])
    plt.ylim(bottom=0, top=params['ylim'][1])
    plt.xlabel('Volume, v'); plt.ylabel('Number Density, f(v,t)')
    plt.title(f'Breakage {case_type.capitalize()}: PINN vs Analytic ({loss_scaling})')
    plt.legend(); plt.grid(True, which='both', ls='--', lw=0.5)
    plt.tight_layout(); plt.show()

    plt.figure(figsize=(10, 6))
    plt.semilogy(pinn.train_loss_history, label='Total Weighted Loss', alpha=0.7)
    plt.semilogy(pinn.data_loss_history, label='Data Loss', linestyle='--', alpha=0.8)
    plt.semilogy(pinn.physics_loss_history, label='Physics Loss', linestyle=':', alpha=0.8)
    plt.xlabel('Iteration'); plt.ylabel('Loss (Log Scale)')
    plt.title(f'Loss History for Breakage {case_type.capitalize()} ({loss_scaling})')
    plt.legend(); plt.grid(True, which='both', ls='--'); plt.tight_layout(); plt.show()

if __name__ == "__main__":
    # RECOMMENDED: Use 'adaptive_huber' for best ADAM performance
    run_test_case(case_type='case1', optimizer='tfp', loss_scaling='adaptive_huber')
    run_test_case(case_type='case2', optimizer='tfp', loss_scaling='adaptive_huber')
    run_test_case(case_type='case3', optimizer='tfp', loss_scaling='adaptive_huber')
    run_test_case(case_type='case4', optimizer='tfp', loss_scaling='adaptive_huber')


    # Compare different loss scaling strategies:
    # run_test_case(case_type='case1', optimizer='tfp', loss_scaling='original')
    # run_test_case(case_type='case1', optimizer='tfp', loss_scaling='huber')
    # run_test_case(case_type='case1', optimizer='tfp', loss_scaling='adaptive_epsilon')
    # run_test_case(case_type='case1', optimizer='tfp', loss_scaling='percentile')
```
